### Data Cleaning: Completed

## 3. Feature Engineering

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month_Name'] = df['Order Date'].dt.strftime("%B")
df['No_of_days'] = (df['Ship Date'] - df['Order Date']).dt.days     # Convert timedelta to number of days
df.loc[df['No_of_days'] < 0, 'No_of_days'] = np.nan
df['Cost'] = df['Sales'] - df['Profit']     # Cost price = Sales - Profit
df['Profit_rate'] = df['Profit'] / df['Cost'].replace(0, np.nan)
df.loc[np.isinf(df['Profit_rate']), 'Profit_rate'] = np.nan
```

```python
df.head()
```

## 4. Pre-compute aggregations

```python
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
category_sales = df.groupby('Category')['Sales'].sum().reset_index()
subcategory_sales = (df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).reset_index())
region_sales = df.groupby('Region')['Sales'].sum().reset_index()
top_customers = df.groupby('Customer Name')['Sales'].sum().nlargest(10).reset_index()
```

## 5. Save Cleaned File

```python
df.to_csv("Superstore_Cleaned.csv", index=False)
print("Cleaned data saved as 'Superstore_Cleaned.csv'")
```

## 6. Data Visualization

### A) Distributions Section 

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Data Distributions', fontsize=16, fontweight='bold')

sns.histplot(df['No_of_days'], bins=10, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Distribution of Shipping Time")
axes[0, 0].set_xlabel("Days to Ship")

sns.histplot(df['Quantity'], bins=20, kde=True, ax=axes[0, 1])
axes[0, 1].set_title("Distribution of Quantity")

sns.histplot(df['Cost'], bins=20, kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Distribution of Cost")

sns.histplot(df['Profit_rate'], bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title("Distribution of Profit Rate")

plt.tight_layout()
plt.show()
```

**Insights**

1. **Distribution of Shipping Time**

- Most orders are shipped within **4–5 days**.
- Very few orders take **0–1 days**, indicating limited same-day/next-day delivery.
- Shipping duration is consistent, with no extreme delays.

2. **Distribution of Quantity**

- Most orders are for **2–4 units**, showing typical small-order retail behavior.
- Larger quantities (>7 units) are rare.
- Pattern suggests a **high volume of small, frequent purchases**.

3. **Distribution of Cost**

- Most products have low cost.
- Cost distribution is **heavily right-skewed**.
- A few high-cost items exist as outliers.

4. **Distribution of Profit Rate**

- Profit rate varies widely across products.
- Most orders show moderate profit **(0.2–0.6 range)**.
- Negative profit rates indicate occasional losses.

### B) Sales Analysis Section

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Sales Analysis', fontsize=16, fontweight='bold')

# Monthly Sales Trend (no sorting needed because it's a line plot)
sns.lineplot(data=monthly_sales, x='Month', y='Sales', hue='Year', ax=axes[0, 0])
axes[0, 0].set_title("Monthly Sales Trend")

# ---- SORTING BEFORE BAR PLOTS ----

# Sort Category Sales
category_sales_sorted = category_sales.sort_values("Sales", ascending=False)

# Sort Region Sales
region_sales_sorted = region_sales.sort_values("Sales", ascending=False)

# Sort Top Customers
top_customers_sorted = top_customers.sort_values("Sales", ascending=False)

# ----------------------------------

# Sales by Category (Bar)
sns.barplot(data=category_sales_sorted, x='Category', y='Sales', ax=axes[0, 1])
axes[0, 1].set_title("Sales by Category")
axes[0, 1].tick_params(axis='x', rotation=45)

# Sales by Region
sns.barplot(data=region_sales_sorted, x='Region', y='Sales', ax=axes[1, 0])
axes[1, 0].set_title("Sales by Region")
axes[1, 0].tick_params(axis='x', rotation=45)

# Top 10 Customers
sns.barplot(data=top_customers_sorted, y='Customer Name', x='Sales', ax=axes[1, 1])
axes[1, 1].set_title("Top 10 Customers by Sales")

plt.tight_layout()
plt.show()
```

**Insights**

1. **Monthly Sales Trend**

- Sales **steadily increase** from **2014 to 2017**.
- **November** is the **strongest** month every year.
- **February–March** consistently show the **lowest sales**.
- **2017** shows the **highest** overall performance.

2. **Sales by Category**

- **Technology leads** sales across all categories.
- **Furniture and Office Supplies** are **almost equal** but below Technology.
- Technology is the primary revenue driver.

3. **Sales by Region**

- **West** region generates the **highest sales**.
- **East** region **closely follows** and performs strongly.
- **Central** region shows **moderate** sales.
- **South** is the **weakest** region with the lowest revenue.

4. **Top 10 Customers by Sales**

- **Sean Miller** is the **top customer** with the highest purchase value.
- **Top customers** contribute significant **sales volume** individually.
- The list shows a **high-value customer segment** that drives revenue.
- Customer spend drops gradually from top to bottom.

### F) Profit Summary

```python
plt.figure(figsize=(8, 5))

# Compute totals
total_sales = df['Sales'].sum()
total_loss = df[df['Profit'] < 0]['Profit'].sum()      # negative value
total_gain = df[df['Profit'] > 0]['Profit'].sum()

# Convert loss to positive magnitude for better visualization
values = [total_sales, abs(total_loss), total_gain]
labels = ['Total Sales', 'Total Loss', 'Total Gain']
colors = ['#1f77b4', '#d62728', '#2ca02c']

sns.barplot(x=labels, y=values, palette=colors)

plt.title("Profit Summary", fontsize=14, fontweight='bold')
plt.ylabel("Amount ($)")
plt.tight_layout()

# Add values on top of bars
for i, v in enumerate(values):
    plt.text(i, v, f"{v:,.0f}", ha='center', va='bottom', fontsize=10)

plt.show()
```

**Insights**

**Profit Summary**

- **Total Sales are significantly high**, indicating strong overall revenue for the business.
- **Total Gain (profit) is positive but much smaller than total sales**, showing that only a portion of revenue converts to actual profit.
- **Total Loss is noticeable**, indicating that certain orders/products are being sold at a loss — likely due to heavy discounts or high costs.
- The presence of losses suggests a need to **review discount strategies, pricing, and cost management**.

## **8. Business Recommendations**


### **1. Optimize Discount Strategy**

* Reduce heavy discounts that lead to negative profit margins.
* Apply targeted discounts only on high-margin items or slow-moving stock.

### **2. Improve Profitability of Furniture Category**

* Revisit pricing, vendor negotiations, or shipping costs for loss-making furniture products.
* Focus on high-performing sub-categories like Chairs and Tables.

### **3. Leverage Seasonal Demand**

* Prepare inventory and marketing campaigns ahead of November, the peak sales month.
* Introduce off-season offers during February–March to balance the sales cycle.

### **4. Strengthen the South Region**

* Investigate low performance through regional market study.
* Increase promotional activities, local partnerships, or logistics improvements.

### **5. Retain High-Value Customers**

* Create loyalty programs or exclusive offers for top-spending customers.
* Identify purchasing patterns to personalize recommendations.

### **6. Enhance Shipping Efficiency**

* Explore faster delivery options to reduce reliance on 4–5 day shipping.
* Improve handling processes to increase next-day or two-day delivery frequency.

### **7. Expand High-Performing Sub-Categories**

* Increase stock and marketing focus on Phones, Chairs, and Storage products.
* Phase out or reconsider low-performing items like Fasteners, Labels, and Envelopes.
